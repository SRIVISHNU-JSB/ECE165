## Question 3

### 3.1 Get the visualization of your state data through day-wise plot.

```python
assigned_states_df = pd.read_csv("~/Downloads/Updated_PowerConsumption_India.csv")
# assigned_states_df

date_df = assigned_states_df['Date']
my_state_df = assigned_states_df['Meghalaya']
neighbor_state_df = assigned_states_df['Tripura']

# plt.xticks(rotation=90)
plt.figure(figsize=(20,15))

plt.subplot(3,1,1)
plt.plot(date_df, my_state_df)

plt.title("Daily Power Consumption of Meghalaya")
plt.xlabel("Date from January 2019 - December 2020")
plt.ylabel("Power in MW (Mega Watts)")

plt.figure(figsize=(20,10))
plt.subplot(3,1,2)
plt.bar(date_df, my_state_df)
plt.bar(date_df, neighbor_state_df)

plt.title("Daily Power Consumption of Both States")
plt.xlabel("Date from January 2019 - December 2020")
plt.ylabel("Power in MW (Mega Watts)")

plt.figure(figsize=(20,10))
plt.subplot(3,1,3)
plt.plot(date_df, neighbor_state_df)

plt.title("Daily Power Consumption of Tripura")
plt.xlabel("Date from January 2019 - December 2020")
plt.ylabel("Power in MW (Mega Watts)")
```

### 3.2 Comment on the change in power consumption pattern, if any, during the lock- down period from April, 2020 to Dec., 2020

## Question 4

### 4.1 Create a record of the sum of the power consumption in every month for both the states.

```python
# summary_df = my_states_df.groupby(['Year', 'month'])[["Meghalaya","Tripura"]].apply(sum).plot(kind='bar')
summary_df = assigned_states_df.groupby(['Year', 'month'])[["Meghalaya","Tripura"]].sum()
summary_df.plot(kind="bar")

# Graph Labels
plt.title("Monthly Power Consumption")
plt.xlabel("Date from January 2019 - December 2020")
plt.ylabel("Power in MW (Mega Watts)")
```

### 4.2 Plot them and detect the months of maximum power and minimum power consumed for both the states.

```python
# summary_df = assigned_states_df.groupby(['Year', 'month'])[["Meghalaya","Tripura"]].sum().plot(kind='bar')
summary_df = assigned_states_df.groupby(['Year', 'month'])[["Meghalaya","Tripura"]].apply(sum)
summary_df.plot(kind="bar")

plt.title("Monthly Power Consumption of Meghalaya & Tripura")

# Maximum and Minimum Power Consumption of Both States
print("Maximum Monthly Power Consumption of Meghalaya")
assigned_states_df[assigned_states_df["Meghalaya"] == assigned_states_df["Meghalaya"].max()]
print("Minimum Monthly Power Consumption of Meghalaya")
assigned_states_df[assigned_states_df["Meghalaya"] == assigned_states_df["Meghalaya"].min()]

print("Maximum Monthly Power Consumption of Tripura")
assigned_states_df[assigned_states_df["Tripura"] == assigned_states_df["Tripura"].max()]
print("Minimum Monthly Power Consumption of Tripura")
assigned_states_df[assigned_states_df["Tripura"] == assigned_states_df["Tripura"].min()]
```

```python
comparison_df = assigned_states_df.groupby(['Year', 'month'])[["Meghalaya","Tripura"]].apply(sum)
# comparison_df1 = assigned_states_df.groupby(['Year', 'month'])[["Meghalaya"]].apply(sum)
# comparison_df2 = assigned_states_df.groupby(['Year', 'month'])[["Tripura"]].apply(sum)

comparison_df.plot(kind="line")

# Graph Labels 
plt.title("Monthly Power Consumption")
plt.xlabel("Date from January 2019 - December 2020")
plt.ylabel("Power in MW (Mega Watts)")
```

### 4.3 Also, detect the mean of the monthly power consumed for both the years separately for the selected states.

```python
# Monthly Mean
# monthly_df = assigned_states_df.groupby(['Year', 'month'])[["Meghalaya","Tripura"]].mean()
# plt.figure()
# monthly_df.plot(kind="bar")

print("Mean Values from the Monthly Power Consumption of Meghalaya")

print("2019 -> ",assigned_states_df[assigned_states_df["Year"] == 2019]["Meghalaya"].mean())
print("2020 -> ",assigned_states_df[assigned_states_df["Year"] == 2020]["Meghalaya"].mean())

print("\nMean Values from the Monthly Power Consumption of Tripura")

print("2019 -> ",assigned_states_df[assigned_states_df["Year"] == 2019]["Tripura"].mean())
print("2020 -> ",assigned_states_df[assigned_states_df["Year"] == 2020]["Tripura"].mean())

# Yearly Mean of both States
yearly_df = assigned_states_df.groupby(['Year'])[["Meghalaya","Tripura"]].mean()
yearly_df.plot(kind="bar")
```
