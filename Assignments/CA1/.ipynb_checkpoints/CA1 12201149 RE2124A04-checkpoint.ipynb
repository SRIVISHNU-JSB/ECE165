```python
# a. Given a circuit in Figure 1, write a Python program to solve it to find out 
# the circuit current 𝑰𝟏, 𝑰𝟐, 𝑰3 and node voltages 𝑽𝒂, 𝑽𝒃

# Set the precision for decimal values with X digits as required
P = 6 # precision, Like 0.000000 and so on

# Resistance in the circuit
R1 = 10 #10 Ohms
R2 = 40 #40 Ohms
R3 = 20 #20 Ohms

## Supply Voltage
Vs = 10 #10 Volts

# Ohms law - V(Voltage)=I(current) x R(Resistance) => I=V/R

# Serial Resistance
Rs = R1
# Parallel Resistance
Rp = ((R2*R3)/(R2+R3))
# Total Resistance
Rt = Rp+Rs
# Total Current It and I1
It = Vs/Rt; I1 = It

# Voltage at Va and Vab
Va = Vs; Vab = I1*R1;
# Voltage at Vb and Vbd
Vb = Va-Vab; Vbd = Vb

# Calculation of Current I1, I2, I3
I1 = Vab/R1
I2 = -(Vb/R3); #I2 = -I2 # For the circuit's current flow is reversed in direction as shown in Figure 1
I3 = Vbd/R2

# Calculation of Current I1, I2, I3
I1 = Vab/R1
I2 = -(Vb/R3); #I2 = -I2 # For the circuit's current flow is reversed in direction as shown in Figure 1
I3 = Vbd/R2

print(f"Current at I1: {round(I1,P)}\nCurrent at I2: {round(I2,P)}\nCurrent at I3: {round(I3,P)}\n\
Voltage at Va: {round(Va,P)}\nVoltage at Vb: {round(Vb,P)}")
```

### Answer

```python
# b. Modify the above code to use function. The function should accept the resister values 𝑹𝟏, 𝑹𝟐 and 𝑹𝟑, 
# as well, the supply voltage 𝑽𝒔 as input parameters passed user.Pass the parameter voltage 𝑽𝒔 using 
# the default value of 10 V. The function should return branch currents 𝑰𝟏, 𝑰𝟐 and node voltages 𝑽𝒂, 𝑽𝒃.

def circuitAnalysis(R1, R2, R3, Vs=10):
    
    # Ohms law - V(Voltage)=I(current) x R(Resistance) => I=V/R

    # Serial Resistance
    Rs = R1
    # Parallel Resistance
    Rp = ((R2*R3)/(R2+R3))
    # Total Resistance
    Rt = Rp+Rs
    # Total Current It and I1
    It = Vs/Rt; I1 = It
    
    # Voltage at Va and Vab
    Va = Vs; Vab = I1*R1;
    # Voltage at Vb and Vbd
    Vb = Va-Vab; Vbd = Vb

    # Calculation of Current I1, I2, I3
    I1 = Vab/R1
    I2 = -(Vb/R3); #I2 = -I2 # For the circuit's current flow is reversed in direction as shown in Figure 1
    I3 = Vbd/R2
    
    return [I1, I2, Va, Vb]

if __name__ == "__main__":
    
    # Note: The input must be separated by space. In case we wish to take input separated by comma (,)
    # we can use it as required by specifying in split() method in List Comprehension.
    R1,R2,R3,Vs=[int(i) for i in input('Enter the values for R1 R2 Rs and Vs: ').split(' ')]

    report = circuitAnalysis(R1,R2,R3,Vs)

    print(f"\nCurrent at I1: {round(report[0],P)}\nCurrent at I2: {round(report[1],P)}\n\
Voltage at Va: {round(report[2],P)}\nVoltage at Vb: {round(report[3],P)}")
    
```

### Q.2 Given a string: “Why is the media here so negative? Why are we in India so embarrassed to recognize our own strengths, our achievements? We are such a great nation. We have so many amazing success stories but we refuse to acknowledge them. Why?” President Abdul Kalam had said in one of his best speeches delivered at Indian Institute of Technology, Hyderabad.

Generate a code to:
#### a. Identify no. of vowels in the string [4]
#### b. Report number of each vowel figuring in the string [4]
#### c. Dig out the vowel with highest number of citation [2]
#### d. Create a user input driven interface for identification of vowels in an input string.[5]

### Answer

```python
# a. Identify no. of vowels in the string

# Instead, we can use casefold() function to lowerCase all the letters as mentioned in this block
vowels = ["A", "E", "I", "O", "U", "a", "e", "i", "o", "u"]

string = "Why is the media here so negative? Why are we in India so embarrassed to recognize our own strengths,\
our achievements? We are such a great nation. We have so many amazing success stories but we refuse to \
acknowledge them. Why?” President Abdul Kalam had said in one of his best speeches delivered at \
Indian Institute of Technology, Hyderabad."

# The function casefold() has been used convert to lowerLetters and to avoid any ignoring cases
# string = string.casefold()

vowel_counter = 0

for character in string:
    if character in vowels:
        vowel_counter += 1

print('Total number of vowels in the given string:',vowel_counter)
```

```python
# Another way a. Identify no. of vowels in the string

def vowelsCounter(string, vowels):
    vowels_found = [each for each in string if each in vowels]
    
    return len(vowels_found)

total_vowels = vowelsCounter(string, vowels);

print('Total number of vowels in the given string:',total_vowels)
```

### Answer

```python
# b. Report number of each vowel figuring in the string
# Create Variables for Vowels Counter and assign to 0
a, e, i, o, u = 0, 0, 0, 0, 0

for char in string:
    if (char in vowels):
        if (char == "A" or char == "a"):
            a += 1
        elif (char == "E" or char == "e"):
            e += 1
        elif (char == "I" or char == "i"):
            i += 1
        elif (char == "O" or char == "o"):
            o += 1
        elif (char == "U" or char == "u"):
            u += 1

print(f"Number of A: {a}\nNumber of E: {e}\nNumber of I: {i}\nNumber of O: {o}\nNumber of U: {u}")
```

### Answer

```python
# c. Dig out the vowel with highest number of citation

# Number of vowels is calculated the above block and storing the number of each vowel in a list
number_of_each_vowels_found = [a, e, i, o, u] 
maximum_citation  = max(number_of_each_vowels_found)

# Find the the place of vowel using index() function
index = number_of_each_vowels_found.index(maximum_citation)
highest_citation_vowel = vowels[index]

print(f"The vowel \'{highest_citation_vowel}\' has maximum number of citations.")
```

### Answer

```python
# d. Create a user input driven interface for identification of vowels in an input string.[5]

# Get the string from user
user_string=input('Enter a string: ')

def SearchForVowels(string, vowels):
    # Create Variables for Vowels Counter and assing to 0
    a, e, i, o, u = 0, 0, 0, 0, 0
    
    vowelsFound = [each for each in string if each in vowels]
    print('\nTotal number of vowels in the given string:',len(vowelsFound))

    for char in string:
        if (char in vowels):
            if (char == "A" or char == "a"):
                a += 1
            elif (char == "E" or char == "e"):
                e += 1
            elif (char == "I" or char == "i"):
                i += 1
            elif (char == "O" or char == "o"):
                o += 1
            elif (char == "U" or char == "u"):
                u += 1
                
    print(f"\nNumber of A: {a}\nNumber of E: {e}\nNumber of I: {i}\nNumber of O: {o}\nNumber of U: {u}")
        
    # Number of vowels is calculated the above block and storing the number of each vowel in a list
    number_of_each_vowels_found = [a, e, i, o, u]
    maximum_citation  = max(number_of_each_vowels_found)

    # Find the the place of vowel using index() function
    index = number_of_each_vowels_found.index(maximum_citation)
    highest_citation_vowel = vowels[index]

    print(f"\nThe vowel \'{highest_citation_vowel}\' has maximum number of citations.")

SearchForVowels(user_string, vowels)
```
