### MatPlotLib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t1 = np.arange(0.0,5.0,0.2)
t2 = np.arange(0.0,5.0,0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, np.exp(-t1),'bo', t2, np.exp(-3*t2), "k")
```

```python
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), "r--")
plt.show()
```

```python
t1 = np.arange(0.0,5.0,0.2)
t2 = np.arange(0.0,5.0,0.02)
t3 = np.arange(0.0,5.0,0.5)
t4 = np.arange(0.0,5.0)

plt.figure()
plt.subplot(221)
plt.plot(t1, np.exp(-t1),'bo', t2, np.exp(-3*t2), "k")

plt.subplot(222)
# plt.plot(t2, np.cos(2*np.pi*t2), "r--")
plt.plot(t2, np.sin(2*np.pi*t2), "r--")
# plt.show()

plt.figure()
plt.subplot(223)
plt.plot(t3, np.cos(2*np.pi*t3), "g:.")

plt.subplot(224)
plt.plot(t4, np.cos(2*np.pi*t4), "y:.")
plt.show()
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9,3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```
