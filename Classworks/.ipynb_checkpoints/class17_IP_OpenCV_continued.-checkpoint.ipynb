```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import cv2
```

```python
img = cv2.imread('/Users/sjsb/Downloads/cameraman.jpg',0)
img.shape
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
img1 = cv2.resize( img, (512, 512))
plt.imshow(img1, cmap = 'gray')
```

```python
imgNeg = img
imgNeg = 255 - img1
plt.imshow(imgNeg, cmap = 'gray')
```

```python
imgTrans = np.transpose(img)
# imgNeg = 255 - img1
plt.imshow(imgTrans, cmap = 'gray')
```

```python
#Horizontal Flip
imgFlip = cv2.flip(img, 1)
imgNeg = 255 - img1
plt.imshow(imgFlip, cmap = 'gray')
```

```python
#Vertical Flip
imgFlip = cv2.flip(img, 0) #0 - for vertiacal and 1 for horizontal
imgNeg = 255 - img1
plt.imshow(imgFlip, cmap = 'gray')
```

```python
#Both Horizontal and Vertical Flip
imgFlip = cv2.flip(img, -1) # -1 - for vertiacal and 1 for horizontal and -1 for both
imgNeg = 255 - img1
plt.imshow(imgFlip, cmap = 'gray')
```
