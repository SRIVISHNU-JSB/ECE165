1. Create a Numpy array of student marks in 5 subjects out of maximum 100 marks. There are in total 40 students.

```python
import numpy as np 
import pandas as pd
```

```python
Marks = np.random.randint(0, 100, (40, 5))
Marks
```

### Q.2. Convert this into a Pandas dataframe corresponding to students ids from 202101 to 202140.


```python
ids = pd.Series( np.arange(202101, 202141) )
ids
```

```python
df = pd.DataFrame( {'ID': ids, 
                  'Sub1': Marks[:,0],
                  'Sub2': Marks[:, 1],
                  'Sub3': Marks[:, 2],
                  'Sub4': Marks[:, 3], 
                'Sub5': Marks[:, 4]}) 
df.head()
```

```python
std_id = np.arange(202101, 202141)
print(std_id.ndim)
print(std_id.shape)
```

```python
std_id = np.reshape(std_id, (40, 1))
print(std_id.ndim)
print(std_id.shape)
```

```python
data = np.concatenate( (std_id, Marks), axis = 1) 
data
```

```python
std_df = pd.DataFrame( data, columns = ['Id', 'SUB1', 'SUB2', 'SUB3', 'SUB4', 'SUB5'])
std_df
```

```python
marks_df = pd.DataFrame( Marks, index = ids, columns = ['S1', 'S2', 'S3', 'S4', 'S5'])
marks_df
```

### 3. Get the average marks in each subject.


```python
std_df["SUB1"].mean()
```

```python
std_df.mean()
```

```python
stdmarks = std_df.drop(columns = 'Id')
stdmarks
```

```python
stdmarks.mean()
```

### 4. Get the percentage of each student and add that for each student in a separate column.

```python
total_marks = std_df['SUB1'] + std_df['SUB2'] + std_df['SUB3'] + std_df['SUB4'] +std_df['SUB5']
total_marks
```

```python
per_marks = total_marks/5
per_marks
```

```python
std_df['Percentage'] = per_marks
std_df
```
