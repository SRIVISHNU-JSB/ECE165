## Introduction to MatPlotLib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [2, 4, 8, 16]
y = [3, 6, 9, 12]
plt.plot(x,y)
```

```python
x = np.arange(0,3,0.1)
y = 3**x + 5

plt.plot(x,y)
```

```python
x = np.arange(0,3,0.1)
y = 3**x + 5

plt.plot(x,y,':')
```

```python
plt.plot(x, x, 'b-', x, x**2, 'c-.', x, x**3, 'm:')
plt.title('Plot Functionality from Matplotlib')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, color = 'k')
plt.legend(labels=['x','x^2','x^3'])
```

```python
plt.plot(x, x, 'b-', x, x**2, 'c-.', x, x**3, 'm:s')
```

```python
plt.plot(x, x, 'b:', x, x**2, 'c-.', x, x**3, 'm:s')
```
